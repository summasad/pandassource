```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "a" :[4,5,6,6],
    "b" :[7,8,9,9],
    "c" :[10,11,12,12],
}, index=list("abcd"))

df
```

```python
# b 컬럼의 값 중 7 이 아닌 values 추출
df[df["b"] != 7]
```

```python
# b 컬럼의 값 중 7 이 아닌 values 추출 + a, b 컬럼만 추출

# df[df["b"] != 7][["a","b"]]

# loc(처리속도 비교적 빠름)
df.loc[df["b"] != 7,["a","b"]]
```

```python
# a 컬럼에 5를 포함하고 있는 값

df[df.a.isin([5])]

# a 컬럼에 5를 포함하지 않는 값
df[~df.a.isin([5])]
```

```python
import numpy as np

df = pd.DataFrame({
    "a":[4,5,6,6,np.nan],
    "b":[7,8,np.nan,9,9],
    "c":[10,11,12,np.nan,12]
}, index=list(range(1,6)))


df
```

```python
# 널 값

df.isnull()

# 특정 컬럼 널 값
df.a.isnull()
df["a"].isnull()
```

```python
df.isnull()
df.isna()
```

```python
# 각 컬럼에 null 의 값이 몇개 존재하느냐?

df.isnull().sum()
```

```python
# 널이 아닌

df.notnull()
```

```python
df.notnull().sum()
```

```python
df["a"].notnull()
```

```python
df.any()
```

```python
#python
4 and 5
True and False

# df[df.b == 7] and df[df.a == 5] : and는 파이썬

# pandas는 &, |, ~
df[(df.b == 7) | (df.a == 5)]
```

#### NaN 처리
- 1) 제거 : dropna()
- 2) 특정한 값으로 대체 : fillna()

```python
df = pd.DataFrame(
    {
        "name" :["Alfred","Batman","Catwoman"],
        "toy" :[np.nan, "Batmobile", "Bullwhip"],
        "born" :[np.nan, '1940-04-25', pd.NaT]
    }
)
df
```

```python
df.isna()

# 널이 존재 시 제거(행)
df.dropna()
```

```python
df.dropna?
```

```python
# 열기준
df.dropna(axis=1)
```

```python
# any : 컬럼 안에 na 하나라도 존재하면 해당하는 행 or 열 제거
# all : na가 모든 행 or 모든 열에 존재해야 제거거
df.dropna(how='all')
```

```python
value = {"name":'noname', "toy":"Bat", "born":"1900-01-01"}

df.fillna(value=value)
```

```python
student_list = {
    "name" : ["John","Nate","Edward","Zara","Wendy","Nate","John"],
    "job" : [
        "teacher","teacher","teacher","student","student","student","student"
    ],
    "age":[40,35,37,25,34,None,None]
}
df = pd.DataFrame(student_list)
df
```

```python
df.info()
```

```python
#
df["age"].fillna(0, inplace=True)
```

```python
# NaN 을 나이의 평균으로 채우기
# mean : 평균, median : 중앙값
# df["age"].fillna(df["age"].mean())

# teacher 는 teacher 평균, student 는 student 평균
df["age"].fillna(df.groupby("job")['age'].transform("median"))
```

```python
data = [
    [np.nan, 2 ,np.nan, 0],
    [3, 4, np.nan, 1],
    [np.nan, np.nan ,np.nan, 5],
]

df = pd.DataFrame(data, columns=list('ABCD'))
df
```

```python
# NaN : 결측치
# 결측치 확인
df.isna()
```

```python
# 결측치 0으로 채우기
df.fillna(0)
```

```python
# 특정 값으로 채우기
# A:0, B:1, C:2, D:3
values = {"A":0, "B":1, "C":2, "D":3}
df.fillna(value=values)
```

```python
# 결측치를 D열의 중앙값으로 채우기
df.fillna(df["D"].median())
```
