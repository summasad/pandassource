```python
seoul_df=pd.read_excel("./resources/.xlsx", usecols='B,D,G,J', skiprows=[0,1,3])
seoul_df.head(3)
```

```python
#   :  , , , 

seoul_df.rename(columns={"(2)":" ", ".1":"", ".2":"", ".3":""}, inplace=True)
seoul_df.columns=[" ", "", "", ""]
```

```python
# cctv_df  seoul_df 
merge_df = pd.merge(seoul_df, cctv_df, on=" ")
```

```python
#   :  
# set_index()

merge_df.set_index(' ', inplace=True)
```

```python
merge_df.head(3)

# cctv  /  * 100 = 1 cctv 
merge_df['1 cctv '] = merge_df['cctv '] / merge_df[""] * 100
```
