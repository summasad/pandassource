#### 원 그래프

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["python", "java", "script", "c#","c++", "etc"]

# startangle : 원하는 각도에서 시작할 수 있으나 각도를 주면 시계 반대 방향으로 그려짐
# counterclock=False : 시계 방향으로 원 작성
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct="%.1f%%")
```

```python
#explode=[0.2,0.1,0,0,0,0]
explode=[0.05]*6

plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct="%.1f%%",explode=explode)
```

```python
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct="%.1f%%",explode=explode)
g=plt.legend(loc=(1.2, 0.3), title="언어별 선호도")
```

```python
colors = ["#ffadad","#ffd6a5","#fdffb6","#caffbf","#9bf6ff","#a0c4ff"]
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct="%.1f%%",explode=explode, colors=colors)
```

```python
wedgeprops={"width":0.8}
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct="%.1f%%",explode=explode, colors=colors, wedgeprops=wedgeprops)
```

```python
wedgeprops={"width":0.8, "edgecolor":"w","linewidth":3}
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct="%.1f%%", colors=colors, wedgeprops=wedgeprops)
```

```python
def custom_autopct(pct):
    return ("%.1f%%" % pct) if pct >=10 else ""

wedgeprops={"width":0.8, "edgecolor":"w","linewidth":3}
plt.pie(values, labels=labels, startangle=90, counterclock=False, autopct=custom_autopct, colors=colors, wedgeprops=wedgeprops)
```

#### 혼합형 그래프
- subplots

```python
# 2행 2열 그래프 작성 -  국어(bar), 영어, 수학(plot), 키(barh), 사회(plot)

fig, axes = plt.subplots(2,2,figsize=(15,10))

# 첫번째 차트
axes[0,0].bar(student_df["이름"],student_df["국어"])
axes[0,0].set_title("학생별 국어 점수")
axes[0,0].legend(["국어점수"])
axes[0,0].grid(linestyle="--",linewidth=0.5)
axes[0,0].set_facecolor("lightyellow")

# 두번째 차트
axes[0,1].plot(student_df["이름"],student_df["영어"])
axes[0,1].plot(student_df["이름"],student_df["수학"])
axes[0,1].set_title("학생별 영어, 수학 점수")
axes[0,1].legend(["영어점수","수학점수"])
axes[0,1].grid(linestyle="--",linewidth=0.5)
axes[0,1].set_facecolor("lightgrey")

# 세번째 차트
axes[1,0].barh(student_df["이름"],student_df["키"])
axes[1,0].set_title("학생별 키")

# 네번째 차트
axes[1,1].plot(student_df["이름"],student_df["사회"])
axes[1,1].set_title("학생별 사회 점수")
axes[1,1].legend(["사회점수"])
axes[1,1].grid(linestyle="--",linewidth=0.5)
```

```python
# 데이터프레임 생성
movie = [
    ["명량",2014, "1761", 8.5],
    ["극한직업",2019, "1626", 9.5],
    ["신과함께",2017, "1441", 9.0],
    ["국제시장",2014, "1425", 8.0],
    ["베테랑",2015, "1341", 8.5],
    ["괴물",2006, "1301", 9.50],
    ["도둑들",2012, "1298", 7.64],
    ["7번방의 선물",2013, "1281", 8.83],
    ["암살",2015, "1270", 9.10],
    ["광해 왕이 된 남자",2012, "1231", 9.25],
]
movie_df=pd.DataFrame(movie,columns=["영화","개봉 연도", "관객수","평점"])
movie_df
```

```python
# 영화 데이터프레임 이용한 차트 작성
# x축은 영화명, y축은 평점인 bar
plt.bar(movie_df["영화"], movie_df["평점"])
```

```python
# 차트 세부설정
# 제목 : 국내 Top 10 영화 평점 정보
# x축 레이블명 : 영화
# y축 레이블명 : 평점
# x축 영화명은 45도 각도로 설정
plt.bar(movie_df["영화"], movie_df["평점"])
plt.title("국내 Top 10 영화 평점 정보")
plt.xlabel("영화")
plt.ylabel("평점")
plt.xticks(rotation=45)
```

```python
# 개봉 연도별 평점 변화 추이를 선 차트로 작성
# 연도별 평균 평점을 구한 뒤 그래프로 작성
# x 축 : 개봉연도, y축 : 평점

# 연도별 평균 평점
# movie_df.groupby("개봉 연도")["평점"].mean() 시리즈 구조

df1 = movie_df.groupby("개봉 연도")["평점"].mean().reset_index()
plt.plot(df1["개봉 연도"], df1["평점"])
```

```python
plt.plot(df1["개봉 연도"], df1["평점"], marker="o")
plt.xticks([2005,2010,2014,2020])
plt.ylim(top=10,bottom=7)
```
